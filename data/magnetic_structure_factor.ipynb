```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as col
plt.style.use(['science'])
```

```python
def sq(arr):
    return arr.reshape((-1,int(np.sqrt(arr.shape[0]))))
```

```python
for t in [240,250,260,270,280,290]:
    arr = np.loadtxt("msf_data/%dK_averaged.txt"%t)
    #arr = np.loadtxt("XMCD#0000..txt.dat")
    #arr = arr[arr[:,0].argsort()]
    qx = sq(arr[:,1])
    qy = sq(arr[:,2])
    re = sq(arr[:,3])
    #im = sq(arr[:,4])
    plt.figure(dpi = 300)
    plt.gca().set_aspect('equal')
    #plt.title("real part")
    plt.xlabel("$q_x$") 
    plt.ylabel("$q_y$")
    plt.pcolormesh(qx,qy,re,cmap='jet',vmax=15)
    plt.colorbar()
    plt.savefig("../figs/msf_experiment_%dK.png"%(t), bbox_inches='tight')
```
